```python
from collections import deque

def is_infected(X, Y, D):
    N = len(X)  

    
    infected = [False] * N

   
    queue = deque()
    queue.append(0)  
    infected[0] = True  

    while queue:
        current_person = queue.popleft()

        
        for i in range(N):
            if not infected[i] and distance(X[current_person], Y[current_person], X[i], Y[i]) <= D:
                infected[i] = True
                queue.append(i)

    return infected



def distance(x1, y1, x2, y2):
    return ((x1 - x2) ** 2 + (y1 - y2) ** 2) ** 0.5



N,D = map(int,input().split())
X = []
Y = []
for _ in range(N):
    x, y = map(int, input().split())
    X.append(x)
    Y.append(y)



infected = is_infected(X, Y, D)


for i in range(N):
    if infected[i]:
        print("Yes")
    else:
        print("No")
```
