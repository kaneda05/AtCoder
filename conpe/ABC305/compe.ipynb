```python
n = int(input())
waters = [5*i for i in range(22)]
min_d = float('inf')
for water in waters:
    d = abs(n-water)
    if d < min_d:
        min_d = d
        ans = water

print(ans)
```

# B

```python
def calculate_distance(p, q):
    distances = {'A': 0, 'B': 3, 'C': 4, 'D': 8, 'E': 9, 'F': 14, 'G': 23}
    return abs(distances[q] - distances[p])

# 入力例
p,q = input().split()

result = calculate_distance(p, q)
print(result)
```
